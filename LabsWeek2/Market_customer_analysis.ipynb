```python
import pandas as pd
```

```python
df1 = pd.read_csv("../Downloads/Data_Marketing_Customer_Analysis_Round2.csv")
df1
```

```python
# Standardizing header names
column_names = df1.columns
column_names_clean = [column.lower().replace(" ", "_") for column in column_names]
df1.columns = column_names_clean
df1.info()
```

```python
df1 = df1.drop(columns = ["unnamed:_0", "customer"])
df1
```

```python
print(df1["gender"]. value_counts())
```

```python
df1.state.unique()
```

```python
df1=df1.drop_duplicates()
df1.info()
```

```python
df1 = df1.reset_index()
df1
```

```python
df1.drop(columns = "index", inplace = True)
df1
```

```python
df1["number_of_open_complains"]=df1["number_of_open_complaints"].fillna(value=0)
df1
```

```python

mean_months = round(df1["months_since_last_claim"].mean())
df1["months_since_last_claim"] = df1 ["months_since_last_claim"].fillna(value=mean_months)
df1.info()
```

```python
def region(x):
    
    if x !=x:
        return x

    else:
        dct_ = {"California": "West Region",
           "Washington": "East" ,
           "Oregon": "North West" ,
            "Arizona": "Central" ,
            "Nevada": "Central"}
    
        return dct_[x]
    
```

```python
df1["region"]= df1["state"].apply(region)
df1
```

```python
list(df1.columns)
```

```python
df1["date_column"]=pd.to_datetime(df1["effective_to_date"])
df1["date_column"].value_counts()
```

```python
df1["month_column"]=df1["date_column"].apply(lambda x:x.month)
df1["month_column"].value_counts()
```
