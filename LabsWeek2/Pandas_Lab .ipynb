```python
# turn "Customer lifetime Value" Column into a float by taking the the last lette % out

#x = value in the columns

def funct1(x):
# here we make sure it is a string
    x = str(x)
    
    if x != "nan":
# here we are selecting all the values of the string except for the last one -> which is the one we want to get rid of i.e. %
        y = x[: -1]
        return float(y)
    else:
        return 0

# here we are applyting the above defined function on any value in the customer_lifetime_value. First we are copying the df to the df2
df2 = df.copy()
df2["customer_lifetime_value"] = df2 ["customer_lifetime_value"].apply(funct1)
```

```python
print(df2["customer_lifetime_value"].value_counts())
```

```python
# turn "number_of_open_complaints" Column into a float -> by selecting the 3 letter of the value only
def funct2(x):
    
    x = str(x)
    
    if x.lower() == "nan":
        return 0
    else:
# here we are selecting only the 3 value of the string -> seems to be the only one showing the amount of open complaints
        return (int(x[2]))


# again we are creating a new data frame and copy the df2 as the begining
df3 = df2.copy()
df3["number_of_open_complaints"] = df3 ["number_of_open_complaints"].apply(funct2)
```

```python
# Transormatings states name into abbreviations
def state_cl(x):

# why can you do it in the below way to check for none value, coz they have the wierd value that when you equal none values, it results a false
# then i know that x is NaN
    if not x == x:
        return x

# here I do the filtering/overwritting
    else:
        dct_ = {"Washington": "WA",
           "Arizona": "AZ" ,
           "Nevada": "NV" ,
           "California": "CA" ,
            "Oregon": "OR",
           "Cali" : "CA",
           "AZ": "AZ",
           "WA" : "WA"}
    
        return dct_[x]

state_cl("Nevada")
```

```python
# standardizing the Gender Column by changing its values to corrects abbrivations

def gender_cl(x):


# why can you do it in the below way to check for none value, coz they have the wierd value that when you equal none values, it results a false
# then i know that x is NaN
    if not x == x:
        return "n"

# here I do the filtering/overwritting
    else:
        dct_ = {"Male": "M",
           "female": "F" ,
           "Femal": "F" ,
            "F": "F" ,
            "M": "M"}
    
        return dct_[x]

gender_cl("Femal")
```

```python
# Bucketing the data - Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central

# creating a new column

df3["Region"] = [" "]*df3.shape[0]
df3.head()
```

```python
# mapping the St abbreviations to the new column called region

def region(x):
    
    if x !=x:
        return x

    else:
        dct_ = {"CA": "West Region",
           "WA": "East" ,
           "OR": "North West" ,
            "AZ": "Central" ,
            "NV": "Central"}
    
        return dct_[x]
    
region("CA")
```

```python
df3["region"] = df3["st"].apply(region)
df3
```
